## Feature creating

### Number of composers

```python
def composer_count(x):
    if x == 'no_composer':
        return 0
    else:
        return sum(map(x.count, ['|', '/', '\\', ';'])) + 1

train['composer'].fillna('no_composer',inplace=True)
test['composer'].fillna('no_composer',inplace=True)
train['composer_count'] = train['composer'].apply(composer_count).astype(np.int8)
test['composer_count'] = test['composer'].apply(composer_count).astype(np.int8)
```

### Number of lyricist

```python
 def lyricist_count(x):
    if x == 'no_lyricist':
        return 0
    else:
        return sum(map(x.count, ['|', '/', '\\', ';'])) + 1
    return sum(map(x.count, ['|', '/', '\\', ';']))

train['lyricist'].fillna('no_lyricist',inplace=True)
test['lyricist'].fillna('no_lyricist',inplace=True)
train['lyricists_count'] = train['lyricist'].apply(lyricist_count).astype(np.int8)
test['lyricists_count'] = test['lyricist'].apply(lyricist_count).astype(np.int8)
```

### Number of genre_ids

```python
type(train['genre_ids'][0])
```

```python
def genre_id_count(x):
    if x == 'no_genre_id':
        return 0
    else:
        return str(x).count('|') + 1
    
train['genre_ids'].fillna('no_genre_id',inplace=True)
test['genre_ids'].fillna('no_genre_id',inplace=True)
train['genre_ids_count'] = train['genre_ids'].apply(genre_id_count).astype(np.int8)
test['genre_ids_count'] = test['genre_ids'].apply(genre_id_count).astype(np.int8)
```

### Whether composer and lyricist are the same

```python
train['artist_composer'] = (train['artist_name'] == train['composer']).astype(np.int8)
train['artist_composer'] = (test['artist_name'] == test['composer']).astype(np.int8)
```

### Whether the song's composer is top 10%

```python
train['msno'].isnull().sum()
```

```python
test['msno'].isnull().sum()
```

```python
def composer_top(x):
    y = x.groupby('composer')['msno'].count().sort_values(ascending = False).reset_index()
    composer_top = list(y.iloc[1:int(len(x)*0.1) ]['composer'])
    x['composer_top'] = x['composer'].isin(composer_top).astype(np.int8)
    return x
train = composer_top(train)
test = composer_top(test)
```

### Whether the song's lyricist is top 10%

```python
def lyricist_top(x):
    y = x.groupby('lyricist')['msno'].count().sort_values(ascending = False).reset_index()
    lyricist_top = list(y.iloc[0:int(len(x)*0.1)]['lyricist'])
    x['lyricist_top'] = x['lyricist'].isin(lyricist_top).astype(np.int8)
    return x
train = lyricist_top(train)
test = lyricist_top(test)
```

### Whether the song's artist is top 30%

```python
def artist_top(x):
    y = x.groupby('artist_name')['msno'].count().sort_values(ascending = False).reset_index()
    artist_top = list(y.iloc[1:int(len(x)*0.3)]['artist_name'])
    x['artist_top'] = x['artist_name'].isin(artist_top).astype(np.int8)
    return x

train = artist_top(train)
test = artist_top(test)
```

### Whether the song's language is top 3

```python
def language_top(x):
    y = x.groupby('language')['msno'].count().sort_values(ascending = False).reset_index()
    language_top = list(y.iloc[0:2]['language'])
    x['language_top'] = x['language'].isin(language_top).astype(np.int8)
    return x

train = language_top(train)
test = language_top(test)
```

### Whether the song contains the genre within top50

```python
def genre_top(x):
    y = x.groupby('genre_ids')['msno'].count().sort_values(ascending = False).reset_index()
    genre_top = list(y.iloc[0:49]['genre_ids'])
    x['genre_top'] = x['genre_ids'].isin(genre_top).astype(np.int8)
    return x

train = genre_top(train)
test = genre_top(test)
```

### Whether the song is from my library in source system tab

One interesting thing is that, if the song is from **my library** then it is more likely to be replayed within a month and if it is from radio then it is less likely. My library is where the user stores their songs locally and hence they really love that song, leading to high mean_target.
On contrary, radio is a random shuffle of songs and hence the user likeability is not predefined leading to low mean_target.

```python
def isin_mylib(x):
    y = np.array(x['source_system_tab'] == 'my library').astype(np.int8)
    x['isin_mylib'] = y
    return x

train = isin_mylib(train)
test = isin_mylib(test)
```

### Whether the length of the song is in the range of $[-\sigma + mean, mean + \sigma]$

```python
def isin_length(x):
    mean = x['song_length'].mean()
    std = x['song_length'].std()
    y = np.array((x['song_length'] <= mean+std) & (x['song_length'] >= mean-std)).astype(np.int8)
    x['isin_length'] = y
    return x

train = isin_length(train)
test = isin_length(test)
```

### Whether the artisit and the lyricist are the same

```python
train['artist_lyricist'] = (train['artist_name'] == train['lyricist']).astype(np.int8)
train['artist_lyricist'] = (test['artist_name'] == test['lyricist']).astype(np.int8)
```
