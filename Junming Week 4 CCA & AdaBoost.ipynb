```python
train = df_train.drop(columns=['msno', 'song_id', 'source_system_tab', 'source_screen_name','source_type','gender','genre_ids'
                               ,'artist_name', 'composer', 'lyricist','name_x', 'isrc_x', 'name_y', 'isrc_y', 'is_train',])
test = df_test.drop(columns=['msno', 'song_id', 'source_system_tab', 'source_screen_name','source_type','gender','genre_ids'
                            ,'artist_name', 'composer', 'lyricist','name_x', 'isrc_x', 'name_y', 'isrc_y', 'is_train',])
```

```python
Train = train.as_matrix()
```

```python
Train.shape
```

```python
Tag = df_train_tag.as_matrix()
```

```python
Tag.shape
```

# CCA

```python
from sklearn.cross_decomposition import CCA

cca = CCA(n_components=20)
cca.fit(Train, Tag)


X_c, Y_c = cca.transform(Train, Tag)
```

```python
X_c.shape
```

```python
Y_c.shape
```

```python
Tag.reshape(-1).shape
```

# Fitting and Evaluating: AdaBoost

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier

clf = AdaBoostClassifier(n_estimators=100)
scores = cross_val_score(clf, X_c, Tag.reshape(-1))
scores.mean()    
```

# Fitting and Evaluating: AdaBoost Without CCA

```python
conmmon = AdaBoostClassifier(n_estimators=100)
scores = cross_val_score(conmmon, Train, Tag.reshape(-1))
scores.mean()    
```
