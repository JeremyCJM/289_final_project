### source_system_tab: fill nan by "others"

```python
print(df_train.source_system_tab.nunique())
print(df_test.source_system_tab.nunique())

print(df_train.source_system_tab.unique())
print(df_test.source_system_tab.unique())
```

```python
df_train.source_system_tab = df_train.source_system_tab.fillna('others')
df_test.source_system_tab = df_test.source_system_tab.fillna('others')
```

### source_screen_name

```python
print(df_train.source_screen_name.nunique())
print(df_test.source_screen_name.nunique())

print(df_train.source_screen_name.unique())
print(df_test.source_screen_name.unique())
```

```python
#--- common values ---
print("common values: \n", set(list(df_train.source_screen_name.unique())) & set(list(df_test.source_screen_name.unique())))

#--- different values ---
print("diff values: \n", set(list(df_train.source_screen_name.unique())) ^ set(list(df_test.source_screen_name.unique())))
```

```python
df_train.source_screen_name = df_train.source_screen_name.fillna('other_sources')
df_test.source_screen_name = df_test.source_screen_name.fillna('other_sources')

df_test['source_screen_name'] = df_test['source_screen_name'].replace(['People local', 'People global'], 'other_sources')
```

### source_type

```python
print(df_train.source_type.nunique())
print(df_test.source_type.nunique())

print(df_train.source_type.unique())
print(df_test.source_type.unique())
```

```python
set(list(df_train.source_type.unique())) ^ set(list(df_test.source_type.unique()))
```

```python
df_train.source_type = df_train.source_type.fillna('other_types')
df_test.source_type = df_test.source_type.fillna('other_types')
```

### gender

```python
df_test.gender = df_test.gender.fillna('unknown')
df_train.gender = df_train.gender.fillna('unknown')
```

### song length: fill nan with mean

```python
df_train['song_length'].fillna((df_train['song_length'].mean()), inplace=True)
df_test['song_length'].fillna((df_test['song_length'].mean()), inplace=True)
```

### language: impute missing with 0

```python
df_train.language = df_train.language.fillna(0)
df_test.language = df_test.language.fillna(0)
```

# Creating new features for artist_name

### artist_name: fill missing with "no_artist"

```python
df_train.artist_name = df_train.artist_name.fillna('no_artist')
df_test.artist_name = df_test.artist_name.fillna('no_artist')
```

## feature 1: whether the featured artist has performed or not

```python
def is_featured(x):
    if ('feat.' in x or 'featuring' in x):
        return 1
    else:
        return 0

df_train['artist_is_featured'] = df_train['artist_name'].apply(is_featured)
df_test['artist_is_featured'] = df_test['artist_name'].apply(is_featured)
```

```python
df_train['artist_is_featured'].value_counts()
```

## NEW whether the featured  composer has performed or not

```python
df_train['composer_is_featured'] = df_train['composer'].apply(str).apply(is_featured)
df_test['composer_is_featured'] = df_test['composer'].apply(str).apply(is_featured)
```

```python
df_train['composer_is_featured'].value_counts()
```

## NEW whether the featured lyricist has performed or not

```python
df_train['lyricist_is_featured'] = df_train['lyricist'].apply(str).apply(is_featured)
df_test['lyricist_is_featured'] = df_test['lyricist'].apply(str).apply(is_featured)
```

```python
df_train['lyricist_is_featured'].value_counts()
```

## feature 2: artist_count

```python
def artist_count(x):
    if x == 'no_artist':
        return 0
    else:
        return sum(map(x.count, ['&', 'and', 'feat.', 'featuring'])) + 1

df_train['artist_count'] = df_train['artist_name'].apply(artist_count).astype(np.int8)
df_test['artist_count'] = df_test['artist_name'].apply(artist_count).astype(np.int8)
```

## NEW composer_count

```python
def composer_count(x):
    if x == 'no_composer':
        return 0
    else:
        return sum(map(x.count, ['&', 'and', 'feat.', 'featuring'])) + 1

df_train['composer_count'] = df_train['composer'].apply(str).apply(composer_count).astype(np.int8)
df_test['composer_count'] = df_test['composer'].apply(str).apply(composer_count).astype(np.int8)
```

## NEW lyricist_count

```python
def lyricist_count(x):
    if x == 'no_lyricist':
        return 0
    else:
        return sum(map(x.count, ['&', 'and', 'feat.', 'featuring'])) + 1

df_train['lyricist_count'] = df_train['lyricist'].apply(str).apply(lyricist_count).astype(np.int8)
df_test['lyricist_count'] = df_test['lyricist'].apply(str).apply(lyricist_count).astype(np.int8)
```

## NEW genre_ids_count

```python
def genre_ids_count(x):
    if x == 'no_genre_ids':
        return 0
    else:
        return sum(map(x.count, ['&', 'and', 'feat.', 'featuring', '|'])) + 1

df_train['genre_ids_count'] = df_train['genre_ids'].apply(str).apply(genre_ids_count).astype(np.int8)
df_test['genre_ids_count'] = df_test['genre_ids'].apply(str).apply(genre_ids_count).astype(np.int8)
```

```python
df_train['genre_ids_count'].value_counts()
```

## feature 3: artist_composer: whether the same

```python
df_train['artist_composer'] = (df_train['artist_name'] == df_train['composer']).astype(np.int8)
df_test['artist_composer'] = (df_test['artist_name'] == df_test['composer']).astype(np.int8)
```

## NEW feature: artist_lyricist: whether the same

```python
df_train['artist_lyricist'] = (df_train['artist_name'] == df_train['lyricist']).astype(np.int8)
df_test['artist_lyricist'] = (df_test['artist_name'] == df_test['lyricist']).astype(np.int8)
```

## NEW feature: composer_lyricist

```python
df_train['composer_lyricist'] = (df_train['composer'] == df_train['lyricist']).astype(np.int8)
df_test['composer_lyricist'] = (df_test['composer'] == df_test['lyricist']).astype(np.int8)
```

## feature 4: artist_composer_lyricist: whether the same

```python
df_train['artist_composer_lyricist'] = ((df_train['artist_name'] == df_train['composer']) & (df_train['artist_name'] == df_train['lyricist']) & (df_train['composer'] == df_train['lyricist'])).astype(np.int8)
df_test['artist_composer_lyricist'] = ((df_test['artist_name'] == df_test['composer']) & (df_test['artist_name'] == df_test['lyricist']) & (df_test['composer'] == df_test['lyricist'])).astype(np.int8)
```

## feature 5: num times a song has been played

```python
_dict_count_song_played_train = {k: v for k, v in df_train['song_id'].value_counts().iteritems()}
_dict_count_song_played_test = {k: v for k, v in df_test['song_id'].value_counts().iteritems()}
def count_song_played(x):
    try:
        return _dict_count_song_played_train[x]
    except KeyError:
        try:
            return _dict_count_song_played_test[x]
        except KeyError:
            return 0
    

df_train['count_song_played'] = df_train['song_id'].apply(count_song_played).astype(np.int64)
df_test['count_song_played'] = df_test['song_id'].apply(count_song_played).astype(np.int64)
```

## feature 6: num times the artist has been played

```python
_dict_count_artist_played_train = {k: v for k, v in df_train['artist_name'].value_counts().iteritems()}
_dict_count_artist_played_test = {k: v for k, v in df_test['artist_name'].value_counts().iteritems()}
def count_artist_played(x):
    try:
        return _dict_count_artist_played_train[x]
    except KeyError:
        try:
            return _dict_count_artist_played_test[x]
        except KeyError:
            return 0

df_train['count_artist_played'] = df_train['artist_name'].apply(count_artist_played).astype(np.int64)
df_test['count_artist_played'] = df_test['artist_name'].apply(count_artist_played).astype(np.int64)
```
