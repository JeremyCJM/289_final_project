## feature 2: artist_count

```python
def artist_count(x):
    if x == 'no_artist':
        return 0
    else:
        return sum(map(x.count, ['&', 'and', 'feat.', 'featuring'])) + 1

df_train['artist_count'] = df_train['artist_name'].apply(artist_count).astype(np.int8)
df_test['artist_count'] = df_test['artist_name'].apply(artist_count).astype(np.int8)
```

## NEW composer_count

```python
def composer_count(x):
    if x == 'no_composer':
        return 0
    else:
        return sum(map(x.count, ['&', 'and', 'feat.', 'featuring'])) + 1

df_train['composer_count'] = df_train['composer'].apply(str).apply(composer_count).astype(np.int8)
df_test['composer_count'] = df_test['composer'].apply(str).apply(composer_count).astype(np.int8)
```

## NEW lyricist_count

```python
def lyricist_count(x):
    if x == 'no_lyricist':
        return 0
    else:
        return sum(map(x.count, ['&', 'and', 'feat.', 'featuring'])) + 1

df_train['lyricist_count'] = df_train['lyricist'].apply(str).apply(lyricist_count).astype(np.int8)
df_test['lyricist_count'] = df_test['lyricist'].apply(str).apply(lyricist_count).astype(np.int8)
```

```python
def genre_id_count(x):
    if x == 'no_genre_id':
        return 0
    else:
        return str(x).count('|') + 1
    
df_train['genre_ids'].fillna('no_genre_id',inplace=True)
df_test['genre_ids'].fillna('no_genre_id',inplace=True)
df_train['genre_ids_count'] = df_train['genre_ids'].apply(genre_id_count).astype(np.int8)
df_test['genre_ids_count'] = df_test['genre_ids'].apply(genre_id_count).astype(np.int8)

def composer_top(x):
    y = x.groupby('composer')['msno'].count().sort_values(ascending = False).reset_index()
    composer_top = list(y.iloc[1:int(len(x)*0.1) ]['composer'])
    x['composer_top'] = x['composer'].isin(composer_top).astype(np.int8)
    return x
df_train = composer_top(df_train)
df_test = composer_top(df_test)

def lyricist_top(x):
    y = x.groupby('lyricist')['msno'].count().sort_values(ascending = False).reset_index()
    lyricist_top = list(y.iloc[0:int(len(x)*0.1)]['lyricist'])
    x['lyricist_top'] = x['lyricist'].isin(lyricist_top).astype(np.int8)
    return x
df_train = lyricist_top(df_train)
df_test = lyricist_top(df_test)

def artist_top(x):
    y = x.groupby('artist_name')['msno'].count().sort_values(ascending = False).reset_index()
    artist_top = list(y.iloc[1:int(len(x)*0.3)]['artist_name'])
    x['artist_top'] = x['artist_name'].isin(artist_top).astype(np.int8)
    return x

df_train = artist_top(df_train)
df_test = artist_top(df_test)

def language_top(x):
    y = x.groupby('language')['msno'].count().sort_values(ascending = False).reset_index()
    language_top = list(y.iloc[0:2]['language'])
    x['language_top'] = x['language'].isin(language_top).astype(np.int8)
    return x

df_train = language_top(df_train)
df_test = language_top(df_test)

def genre_top(x):
    y = x.groupby('genre_ids')['msno'].count().sort_values(ascending = False).reset_index()
    genre_top = list(y.iloc[0:49]['genre_ids'])
    x['genre_top'] = x['genre_ids'].isin(genre_top).astype(np.int8)
    return x

df_train = genre_top(df_train)
df_test = genre_top(df_test)

def isin_mylib(x):
    y = np.array(x['source_system_tab'] == 'my library').astype(np.int8)
    x['isin_mylib'] = y
    return x

df_train = isin_mylib(df_train)
df_test = isin_mylib(df_test)
```
