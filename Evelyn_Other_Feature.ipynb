# Creating new features

```python
import numpy as np
import pandas as pd
```

```python
df_train= pd.read_csv("df_train_week2.csv")
df_test= pd.read_csv("df_test_week2.csv")
```

```python
df_test= pd.read_csv("df_test_week2.csv")
```

## New Feature 1: Whether composer and lyricist are the same

```python
df_train['composer_lyricist'] = (df_train['composer'] == df_train['lyricist']).astype(np.int8)
df_test['composer_lyricist'] = (df_test['composer'] == df_test['lyricist']).astype(np.int8)
```

## New Feature 2: Whether contains “()” in artist_name

```python
def find_pa(x):
    return '(' in x[0]


df_train['find_pa_in_artist_name'] = np.apply_along_axis(find_pa, 0, df_train['artist_name'])
df_test['find_pa_in_artist_name'] = np.apply_along_axis(find_pa, 0, df_test['artist_name'])
```

## New Feature 3: Whether contains”()” in lyricists

```python
def find_pa(x):
    return '(' in x


df_train['find_pa_in_lyricists'] = np.apply_along_axis(find_pa, 0, df_train['lyricist'])
df_test['find_pa_in_lyricists'] = np.apply_along_axis(find_pa, 0, df_test['lyricist'])
```

## New Feature 4: Whether contains”()” in composers

```python
def find_pa(x):
    return '(' in x


df_train['find_pa_in_composers'] = np.apply_along_axis(find_pa, 0, df_train['composer'])
df_test['find_pa_in_composers'] = np.apply_along_axis(find_pa, 0, df_test['composer'])
```

## New Feature 5: Whether contains “-” in artist_name

```python
def find_dash(x):
    return '-' in x[0]
    

df_train['find_dash_in_artist_name'] = np.apply_along_axis(find_dash, 0, df_train['artist_name'])
df_test['find_dash_in_artist_name'] = np.apply_along_axis(find_dash, 0, df_test['artist_name'])
```

## New Feature 6: Whether contains “-” in lyricists

```python
def find_dash(x):
    return '-' in x


df_train['find_dash_in_lyricists'] = np.apply_along_axis(find_dash, 0, df_train['lyricist'])
df_test['find_dash_in_lyricists'] = np.apply_along_axis(find_dash, 0, df_test['lyricist'])
```

## New Feature 7: Whether contains “-” in composers

```python
def find_dash(x):
    return '-' in x


df_train['find_dash_in_composers'] = np.apply_along_axis(find_dash, 0, df_train['composer'])
df_test['find_dash_in_composers'] = np.apply_along_axis(find_dash, 0, df_test['composer'])
```

## New Feature 8: num times the composer has been played

```python
_dict_count_composer_played_train = {k: v for k, v in df_train['composer'].value_counts().iteritems()}
_dict_count_composer_played_test = {k: v for k, v in df_test['composer'].value_counts().iteritems()}
def count_composer_played(x):
    try:
        return _dict_count_composer_played_train[x]
    except KeyError:
        try:
            return _dict_count_composer_played_test[x]
        except KeyError:
            return 0

df_train['count_composer_played'] = df_train['composer'].apply(count_composer_played).astype(np.int64)
df_test['count_composer_played'] = df_test['composer'].apply(count_composer_played).astype(np.int64)
```

## New Feature 9: num times the lyricist has been played

```python
_dict_count_lyricist_played_train = {k: v for k, v in df_train['lyricist'].value_counts().iteritems()}
_dict_count_lyricist_played_test = {k: v for k, v in df_test['lyricist'].value_counts().iteritems()}
def count_lyricist_played(x):
    try:
        return _dict_count_lyricist_played_train[x]
    except KeyError:
        try:
            return _dict_count_lyricist_played_test[x]
        except KeyError:
            return 0

df_train['count_lyricist_played'] = df_train['lyricist'].apply(count_lyricist_played).astype(np.int64)
df_test['count_lyricist_played'] = df_test['lyricist'].apply(count_lyricist_played).astype(np.int64)
```

## New Feature 10: registration before 2000

```python
def checktime(x):
    year = int(str(x[0])[:4])
    return year < 2000
df_train['registration_before_2000'] = np.apply_along_axis(checktime, 0, df_train['registration_init_time'])
df_test['registration_after_2000'] = np.apply_along_axis(checktime, 0, df_test['registration_init_time'])
```

```python
def checktime(x):
    year = int(str(x[0])[:4])
    return year >= 2000 and year < 2005
df_train['registration_before_2005'] = np.apply_along_axis(checktime, 0, df_train['registration_init_time'])
df_test['registration_before_2005'] = np.apply_along_axis(checktime, 0, df_test['registration_init_time'])
```
